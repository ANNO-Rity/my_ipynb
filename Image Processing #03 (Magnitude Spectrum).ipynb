```python
import cv2
import numpy as np
from  matplotlib import pyplot as plt
```

```python
img=cv2.imread('/content/drive/MyDrive/Colab Notebooks/gambar_akhir.jpg', 0)
f=np.fft.fft2(img)
fshift=np.fft.fftshift(f)
plt.magnitude_spectrum=2000000*np.log(np.abs(fshift))

#plt.subplot(121),plt.imshow(img,cmap='autumn')
#plt.title('Original Colored'), plt.xticks([]),plt.yticks([])
plt.subplot(121),plt.imshow(img,cmap='gray')
plt.title('Original Grayscale'), plt.xticks([]),plt.yticks([])
plt.subplot(122),plt.imshow(plt.magnitude_spectrum, cmap='rainbow')
plt.title('Magnitude Spectrum'), plt.xticks([]),plt.yticks([])
plt.show()
```

```python
img=cv2.imread('/content/drive/MyDrive/Colab Notebooks/gambar_akhir.jpg')
img = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
dft = np.fft.fft2(img)
dft_shift = np.fft.fftshift(dft)
phase_spectrum = np.angle(dft_shift)

ax1 = plt.subplot(1,2,1)
ax1.imshow(img, cmap='gray')

ax2 = plt.subplot(1,2,2)
ax2.imshow(phase_spectrum, cmap='seismic')

plt.show()
```
