```python
import os
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
from google.colab.patches import cv2_imshow
img=cv2.imread('/content/drive/MyDrive/RosinSR3.jpg', cv2.IMREAD_UNCHANGED)
img2 =cv2.imread('/content/drive/MyDrive/Colab Notebooks/gambar_akhir.jpg', cv2.IMREAD_UNCHANGED)
cv2_imshow(img)
cv2_imshow(img2)
```

```python
from PIL import Image
img = Image.open('/content/drive/MyDrive/RosinSR3.jpg') # Buka gambar yang ingin Anda ubah ukurannya
img = img.resize((1443, 1920), Image.ANTIALIAS) # Ubah ukuran gambar menjadi 1920 x 1080 dengan filter antialias
img.save('image_1080p.jpg') # Simpan gambar yang telah diubah ukurannya
```

```python
import cv2
import numpy as np
img = cv2.imread('image_1080p.jpg') # Baca gambar yang telah diubah ukurannya
kernel = np.array([[-1,-1,-1], [-1,9,-1], [-1,-1,-1]]) # Buat kernel untuk penajaman
img = cv2.filter2D(img, -1, kernel) # Terapkan filter pada gambar
cv2.imwrite('gambar_tajam.jpg', img) # Simpan gambar yang telah dipertajam
```

```python
from PIL import Image
img = Image.open('gambar_tajam.jpg') # Buka gambar yang telah dipertajam
img = img.convert('RGB') # Ubah mode warna gambar menjadi RGB
img.save('gambar_akhir.jpg', format='JPEG', quality=100) # Simpan gambar dengan format JPEG dan kualitas 90
```

```python
plt.subplot(121), plt.imshow(img),plt.title('Original')
plt.xticks([]),plt.yticks([])
plt.subplot(122), plt.imshow(img2),plt.title('Upscalled')
plt.xticks([]),plt.yticks([])
plt.show()
```
